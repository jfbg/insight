## Data Challenge 2

Here I convert all CSV files into pSQL tables in the panama database.

```python
import numpy as np
import pandas as pd
import qgrid 
qgrid.nbinstall(overwrite=True)

from sqlalchemy import create_engine
from sqlalchemy_utils import database_exists, create_database
import psycopg2

# !pythan -m pip install ipython-sql
%load_ext sql
%config SqlMagic.autopandas=True

# %qtconsole
```

```python
data_folder = 'data/offshore_leaks_csvs-20160621/'

# data files:
# Addresses.csv
# Intermediaries.csv
# all_edges.csv
# Entities.csv
# Officers.csv

engine = create_engine('postgresql://JFBG@localhost/panama')
# print(database_exists(engine.url))
```

First start trying to load the data into pandas df. However the files are big and this might become and issue (some are >1M rows). So might have to convert to a pgSQL db.

### ADDRESSES

```python
df = pd.read_csv(data_folder + 'Addresses.csv',low_memory=False)
df.to_sql('addresses', engine, if_exists='replace')
# Add to pSQL database
df.head()
# qgrid.show_grid(df)
```

### Intermediaries


```python
df = pd.read_csv(data_folder + 'Intermediaries.csv',low_memory=False)
df.to_sql('intermediaries', engine, if_exists='replace')
# Add to pSQL database
df.head()
# qgrid.show_grid(df)
```

### ALL EDGES


```python
df = pd.read_csv(data_folder + 'all_edges.csv',low_memory=False)
df.to_sql('all_edges', engine, if_exists='replace')
# Add to pSQL database
df.head()
# qgrid.show_grid(df)
```

### ENTITIES

```python
df = pd.read_csv(data_folder + 'Entities.csv',low_memory=False)
df.to_sql('entities', engine, if_exists='replace')
# Add to pSQL database
df.head()
# qgrid.show_grid(df)
```

### OFFICERS

```python
df = pd.read_csv(data_folder + 'Officers.csv',low_memory=False)
df.to_sql('officers', engine, if_exists='replace')
# Add to pSQL database
df.head()
# qgrid.show_grid(df)
```
